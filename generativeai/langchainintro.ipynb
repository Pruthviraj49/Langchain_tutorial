### Generative AI application using agents with Gemini 

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
response = model.invoke("What is langchain explain me with a simple example so that 10 year old can understand")
print(response.content)
```

```python
print(response.response_metadata)
```
