### Generative AI application using agents with Gemini 

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
response = model.invoke("What is langchain explain me with a simple example so that 10 year old can understand")
print(response.content)
```

```python
print(response.response_metadata)
```

### Groq model integration

```python
import os 
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:llama-3.1-8b-instant")
response = model.invoke("Write a joke for me in hindi")
print(response.content)
```

### Streaming and Batch

```python
model.stream("Write me a story in 500 words")  # Gives response in chunks
```

```python
for chunk in model.stream("Write me a story in 500 words"):
    print(chunk.text, end="|", flush=True)
```

#### Batch (Prallel request serving)

```python
model.batch(["Write me a story in 500 words", "Write me a eassay in 500 words", "Write me a sript in 500 words"],
config={
    'max_concurrency' : 5, # Limit to the parallel calls
})
```
