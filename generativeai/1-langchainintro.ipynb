### Groq model integration

### Streaming and Batch

```python
for chunk in model.stream("Write me a 500 words story"):
    print(chunk.text, end="|", flush= True)
```

Batching acollection of independent requests to the model can significantly improve the performance and reduce costs, as the processing can be done in parallel

```python
model.batch(
    [
        "Tell me about the india",
        "Hoe do airoplane fly?",
       

    ],
    config={
        'max_concurrency': 5, # limit for parallel requests
    }
)
```
