### Agent with Memory

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver

```

```python

agent = create_agent(
    model="groq:qwen/qwen3-32b",
    checkpointer = InMemorySaver()
)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
# Run with thread id
config = {"configurable":{"thread_id": "test1"}}
response = agent.invoke({"messages": [HumanMessage(content= "Hi my name is Pruthvi")]},config = config)
response
```

```python
agent.invoke({"messages": [HumanMessage(content= "Hi What is my name?")]},config = config)
```

```python
config1 = {"configurable":{"thread_id": "test2"}}
response = agent.invoke({"messages": [HumanMessage(content= "Hi my name is Alice")]},config = config1)
agent.invoke({"messages": [HumanMessage(content= "Hi What is my name in one word?")]},config = config1)
```
